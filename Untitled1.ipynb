```python
#IMPORT THE LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sharma-kohli.csv')
df
```

```python
plt.plot(df['index'],df['V Kohli'],color='#FF0000',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(df['index'],df['RG Sharma'],color='#0000FF',linestyle='dashdot',linewidth=2,marker='o')
plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')
plt.legend(loc='upper left')
plt.show()
```
